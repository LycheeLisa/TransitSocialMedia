```python
def get_soup_loaded(url,trials):
    driver = webdriver.Chrome()
#     driver.get(url)
    click=True
    counter=0
    while click==True:
        try:
            driver.get(url)
#             cross = driver.find_element_by_xpath("""//*[@id="root-featured-articles"]/div/div[2]/div/button""")
#             cross.click()
            cross = driver.find_element_by_xpath("""//*[@id="root-latest-reviews"]/dive/dive[2]/div/button""")
            cross.click()
            buttons=driver.find_elements_by_xpath("//button[@title='load-more-button']")
#             buttons = driver.find_elements_by_css_selector("/load-more-button")
            for button in buttons:
                button.click()
                print(button.text.strip())

            counter=counter+1
            print(counter)
            if counter==trials:
                break
        except:
            click=False
    soup = BeautifulSoup(driver.page_source,'html.parser')
    return soup
```

```python
soup=get_soup_loaded('https://www.blogto.com/eat_drink/',trials=10)
```

```python
sth=soup.find_all("p",{"article-thumbnail-title"})
```

```python
mention_list=list()
for gi in sth:
    mention_list.append(gi.span.get_text().strip())
```

```python
BlogTO=pd.DataFrame(columns=["titles"])
BlogTO["titles"]=mention_list
# BlogTO.to_csv("BlogTO_mentions.csv",index=False)
```

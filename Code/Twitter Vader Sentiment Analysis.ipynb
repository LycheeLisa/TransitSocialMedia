```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
def vader_sentiment(tweet):
    try:
        score = analyzer.polarity_scores(tweet)
        return score
    except Exception as e: return e
```

```python
df = pd.read_csv('twitter_df_clean2.csv', index_col=0)
```

```python
vader = df.loc[:,'clean_body'].apply(lambda x: vader_sentiment(x))
df = df.assign(vader=vader.values)
#df.to_csv('twitter_vader_sentiment2.csv')
```

```python
compound = []
pos = []
neu = []
neg = []
for row in df.itertuples():
    c = row.vader.get('compound')
    compound.append(c)
    p = row.vader.get('pos')
    pos.append(p)
    neutral = row.vader.get('neu')
    neu.append(neutral)
    negative = row.vader.get('neg')
    neg.append(negative)

df['compound_sentiment'] = compound
df['positive_sentiment'] = pos
df['negative_sentiment'] = neg
df['neutral_sentiment'] = neu
```

```python
#df.to_csv('twitter_vader_sentiment2.csv')
```

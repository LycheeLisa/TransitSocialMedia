```python
#Importing BeautifulSoup
from bs4 import BeautifulSoup
import requests
import csv
```

```python
#Get source code from website
source = requests.get('https://www.yelp.ca/local_yelp').text

soup = BeautifulSoup(source, 'lxml')

#Confirmed, response object returns from BeautifulSoup. 
```

```python
#CSV Writer (Needs to be before for loop...)
#CSV Writer also has a line at the end of the for loop

filename = "TheLocalYelp.csv"
f = open(filename, 'w')

headers = "restaurant, location"

f.write(headers)

#Find table with restaurant names and locations. 
# For loop for ALL restuarants on the list.
#Indent pulls for for loop


for table in soup.find_all('table'):
    try:
        #Take restaurant name
        restaurant = table.strong.text
        print(restaurant)
        #Take restaurant location
        location = table.find('td', style='color: #333333; padding: 6px 0px; font-size: 14px; font-family: Helvetica Neue, Arial, sans-serif;;').text
        print(location)
    #skip comment tables
    except:
        print('')
    
#Write to 2 columns


    f.write(restaurant +','+ location.replace(',',''))
    
f.close()
```
